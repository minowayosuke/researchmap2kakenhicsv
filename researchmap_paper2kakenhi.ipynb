```python
# 使い方
# 1. このセルを実行（左の三角ボタンを押す）
# 2. 左の三角ボタン→四角ボタンの周りがぐるぐるする→三角ボタンに戻る、までしばらくまつ
# 3. 画面を下にスクロールし、「jsonlアップロード」ボタンをクリック→「choose files」ボタンをクリックして、jsonlファイルをアップロード。100% doneになるまでまつ
# 4. 「変換を実行」ボタンを押して、変換後のファイルをダウンロード

!pip install jsonlines
import jsonlines
import csv
import ipywidgets as widgets
from google.colab import files
from IPython.display import display

def process_jsonl_to_csv(jsonl_file, csv_file):
    with jsonlines.open(jsonl_file, 'r') as reader, open(csv_file, mode='w', encoding='shift_jis', newline='') as csvfile:
        csv_writer = csv.writer(csvfile)
        for obj in reader:
            merge = obj['merge']

            # Get identifier if exists
            identifier = merge['identifiers']['doi'][0] if 'doi' in merge['identifiers'] else ''

            # Get authors
            authors = ','.join([author['name'] for author in merge['authors']['en']])

            # Get paper_title
            paper_title = merge['paper_title']['en']

            # Get publication_name
            publication_name = merge['publication_name']['en']

            # Get volume if exists
            volume = merge['volume'] if 'volume' in merge else ''

            # Get publication_date
            publication_date = merge['publication_date'][:4]

            # Get starting_page and ending_page
            starting_page = merge.get('starting_page', '')
            ending_page = merge.get('ending_page', '')
            pages = f"{starting_page}-{ending_page}" if ending_page else starting_page

            # Get referee value
            referee = 1 if merge.get('referee', False) else 0

            # Write to CSV
            csv_writer.writerow([identifier, authors, paper_title, publication_name, volume, publication_date, pages, referee, 0, 0])

def on_upload_button_click(button):
    uploaded_files = files.upload()
    uploaded_file_name = next(iter(uploaded_files))
    with open('uploaded.jsonl', 'wb') as f:
        f.write(uploaded_files[uploaded_file_name])

def on_convert_button_click(button):
    process_jsonl_to_csv('uploaded.jsonl', 'output.csv')
    files.download('output.csv')

# Upload button
upload_button = widgets.Button(description='jsonlアップロード')
upload_button.on_click(on_upload_button_click)

# Convert button
convert_button = widgets.Button(description='変換を実行', disabled=False)
convert_button.on_click(on_convert_button_click)

# Display buttons
display(upload_button)
display(convert_button)
```
