```python
# 使い方
# 1. このセルを実行（左の三角ボタンを押す）
# 2. 左の三角ボタン→四角ボタンの周りがぐるぐるする→三角ボタンに戻る、までしばらくまつ
# 3. 画面を下にスクロールし、「jsonlアップロード」ボタンをクリック→「choose files」ボタンをクリックして、jsonlファイルをアップロード。アップロードが100% doneとなるまで待つ
# 4. 「変換を実行」ボタンを押して、変換後のファイルをダウンロード

! pip install jsonlines
import jsonlines
import csv
import ipywidgets as widgets
from google.colab import files
from IPython.display import display

def process_jsonl_to_csv(jsonl_file, csv_file):
    with jsonlines.open(jsonl_file, 'r') as reader, open(csv_file, mode='w', encoding='shift_jis', newline='') as csvfile:
        csv_writer = csv.writer(csvfile)
        for obj in reader:
            merge = obj['merge']

            # Get presenters
            presenters = ','.join([p['name'] for p in list(merge['presenters'].values())[0]])

            # Get presentation_title
            presentation_title = list(merge['presentation_title'].values())[0]

            # Get event
            event = list(merge['event'].values())[0]

            # Get from_event_date and to_event_date
            from_event_year = merge['from_event_date'][:4]
            to_event_year = merge['to_event_date'][:4]

            # Get invited flag
            invited = int(merge['invited'])

            # Get languages flag
            languages_flag = int('languages' in merge and 'eng' in merge['languages'])

            # Write to CSV
            csv_writer.writerow([presenters, presentation_title, event, from_event_year, to_event_year, invited, languages_flag])

def on_upload_button_click(button):
    uploaded_files = files.upload()
    uploaded_file_name = next(iter(uploaded_files))
    with open('uploaded.jsonl', 'wb') as f:
        f.write(uploaded_files[uploaded_file_name])

def on_convert_button_click(button):
    process_jsonl_to_csv('uploaded.jsonl', 'kakenhi.csv')
    files.download('kakenhi.csv')

# Upload button
upload_button = widgets.Button(description='jsonlアップロード')
upload_button.on_click(on_upload_button_click)

# Convert button
convert_button = widgets.Button(description='変換を実行', disabled=False)
convert_button.on_click(on_convert_button_click)

# Display buttons
display(upload_button)
display(convert_button)

```
